### 3) Seleccionar el número y nombre de los empleados que ganan más de 1000.

**Tip**: la cláusula *WHERE* se utiliza para filtrar filas específicas de una tabla o conjunto de datos basándose en una condición. La condición se evalúa para cada fila en la tabla, y solo las filas que cumplan con la condición se incluyen en el resultado de la consulta. Para la condición se utiliza: <, <=, >, >=, =, <>

Se pueden realizar consultas compuestas, usando operadores lógicos AND y OR.

El resultado esperado es:

image.png

```
# Consulta
# Seleccionar el número y nombre de los empleados que ganan más de 1000
#
%%sql


SELECT ...
FROM ...
WHERE ...
```

#### Solución:

Puedes acudir a ella en caso de quieras revisar la solución.

```
%%sql

SELECT empno, ename
FROM EMP
WHERE sal > 1000;
```

### 5) Mostrar los nombres de los empleados que ganan más de 1000 y trabajan en el departamento 10, en orden alfabético.

Tip: la cláusula *ORDER BY* te permite ordenar el resultado de la consulta en función de los valores de una columna. Puedes ordenar en orden ascendente (ASC) o descendente (DESC).

El resultado esperado es:

image.png

```
# Consulta
# Mostrar los nombres de los empleados que ganan por lo menos 1000 y trabajan en el departamento 10,
# en orden alfabético.
#
%%sql

SELECT
FROM
WHERE ... AND ...
ORDER BY ...
```

### 7) Mostrar toda la información de los empleados que su nombre empieza por la letra K.

**Tip**: La cláusula "*LIKE*" en SQL se utiliza para buscar patrones de texto en los datos de una columna. Es útil cuando deseas realizar búsquedas que no se basan en un valor exacto, sino en un patrón que coincida con una parte del texto en la columna.

El resultado esperado es:

image.png

```
# Consulta
# Mostrar toda la información de los empleados que su nombre empieza por la letra K
#
%%sql


SELECT ...
FROM EMP
WHERE ename LIKE ...
```

#### Solución:

Puedes acudir a ella en caso de quieras revisar la solución.

```
%%sql

SELECT *
FROM EMP
WHERE ename LIKE 'K%';
```

### 8) Seleccionar los números, nombre y cargos de los empleados que ganan entre 1000 y 2000.

**Tip**: La cláusula "*BETWEEN*" en SQL se utiliza para filtrar filas en una consulta basadas en un rango de valores. Esta cláusula es útil cuando deseas seleccionar registros cuyos valores estén dentro de un intervalo específico en una columna.

El resultado esperado es:

image.png

```
# Consulta
# Seleccionar los números, nombre y cargos de los empleados que ganan entre 1000 y 2000.
#
%%sql

SELECT
FROM ...
WHERE sal BETWEEN .... AND ...
```

#### Solución:

Puedes acudir a ella en caso de quieras revisar la solución.

```
%%sql

SELECT empno, ename, job
FROM EMP
WHERE sal BETWEEN 1000 AND 2000;
```

El resultado espereado es:

image.png

```
# Consulta
# Listado de los cargos existentes, sin repeticiones.
#
%%sql

SELECT distinct ...
FROM ...
```

#### Solución:
